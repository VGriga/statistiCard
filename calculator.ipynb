```python
import random
from collections import Counter
```

```python
deck = full_deck.copy()
gone_cards = []

while len(deck)>20:
    my_cards = random.sample(population=deck, k=2)
    deck.remove(my_cards[0])
    deck.remove(my_cards[1])

    diller_cards = random.sample(population=deck, k=2)
    deck.remove(diller_cards[0])
    deck.remove(diller_cards[1])

    probs = calculator(my_cards, diller_cards, gone_cards)

    gone_cards = gone_cards + my_cards + diller_cards

    print(my_cards)
    if probs == (-1,-1,-1):
        print('looser')
    elif probs == (-2,-2,-2):
        print('winner')
    else:
        print(f'    prob_overtake  = {probs[0]*100:2.2f} %') 
        print(f'    prob_blackjack = {probs[1]*100:2.2f} %') 
        print(f'    prob_undertake = {probs[2]*100:2.2f} %')
        
    while probs[0]<0.4 and probs[0]>0:
        my_cards = my_cards + random.sample(population=deck, k=1)
        deck.remove(my_cards[-1])
        probs = calculator(my_cards, diller_cards, gone_cards)
        gone_cards = gone_cards + [my_cards[-1]]

        print('   ',my_cards)
        if probs == (-1,-1,-1):
            print('   ','looser')
        elif probs == (-2,-2,-2):
            print('   ','winner')
        else:
            print('   ',f'    prob_overtake  = {probs[0]*100:2.2f} %') 
            print('   ',f'    prob_blackjack = {probs[1]*100:2.2f} %') 
            print('   ',f'    prob_undertake = {probs[2]*100:2.2f} %')
            
    print( )
```
